В этой задаче Вам предлагается попробовать стандартную технику локального поиска (local search) в применении к задаче о сбалансированном разбиении графа. Мы будем рассматривать безвесовый вариант задачи с параметром балансировки $\alpha=\frac{1}{2}$:

```python
def read_col_file(filename):
    with open(filename, 'r') as file:
        vertices, edges = set(), set()
        for line in file:
            line = line.strip()
            if line.startswith('p'):
                vertices = set(range(1, int(line.split()[-2]) + 1))
            elif line.startswith('e'):
                edges.add(tuple(map(int, line.split()[-2:])))
        return (vertices, edges)
```

```python
def basic_local_search(graph):
    def edges_between(v1, v2):
        res = 0
        for v in v1:
            for u in v2:
                if M[v][u]:
                    res += 1
        return res
    
    k = 1
    
    Vert = list(graph[0])
    Edges = list(graph[1])
    n = len(Vert)
    M = [[False] * (n + 1) for i in range(n + 1)]
    for e in Edges:
        M[e[0]][e[1]] = True
        M[e[1]][e[0]] = True
    Vert1 = set(Vert[:len(Vert) // 2])
    Vert2 = set(Vert[len(Vert) // 2:])
    while True:
        max_delta = 0
        b_v1 = []
        b_v2 = []
        for v_s in itertools.product(Vert1, repeat=k):
            for u_s in itertools.product(Vert2, repeat=k):
                v1_s = set(v_s)
                v2_s = set(u_s)
                if len(v1_s) != len(v2_s):
                    continue
                Vert1_P = Vert1 - v1_s
                Vert2_P = Vert2 - v2_s
                delta = edges_between(Vert1_P, v2_s) + edges_between(v1_s, Vert2_P) - \
                      edges_between(v1_s, Vert1_P) - edges_between(v2_s, Vert2_P)
                if delta > max_delta:
                    max_delta = delta
                    b_v1 = v1_s
                    b_v2 = v2_s
        if max_delta == 0:
            break
        new_Vert1 = (Vert1 - b_v1) | b_v2
        new_Vert2 = (Vert2 - b_v2) | b_v1
        Vert1 = new_Vert1
        Vert2 = new_Vert2
    print('V1: ', Vert1)
    print('edges between: ', edges_between(Vert1, Vert2))
    return Vert1
```
